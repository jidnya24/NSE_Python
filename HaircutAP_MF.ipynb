```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

# Read data from CSV file
df = pd.read_csv("C:\\Users\\HP\\Downloads\\Haircut_AP\\APPSEC_COLLVAL_01042024.csv")

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b')

def plot_stock_closing_price(stock_name):
    # Filter DataFrame for the given stock name
    stock_df = df[df['StockName'] == stock_name]
    
    # Sort DataFrame by date
    stock_df = stock_df.sort_values(by='Date')

    # Create figure and axis objects
    fig, axs = plt.subplots(2, 1, figsize=(12, 10))

    # Plotting closing price in the first subplot
    axs[0].plot(stock_df['Date'], stock_df['Closing'], color='blue', label='Closing Price', marker='o')
    axs[0].set_xlabel('Date')
    axs[0].set_ylabel('Prev Closing', color='blue')
    axs[0].tick_params('y', colors='blue')

    # Rotate date labels for the first subplot
    axs[0].xaxis.set_major_formatter(DateFormatter('%d-%b'))
    axs[0].tick_params(axis='x', rotation=45)
    
    # Annotate each point with its y-value
    for i, (date, closing) in enumerate(zip(stock_df['Date'], stock_df['Closing'])):
        axs[0].annotate(f'{closing:.2f}', (date, closing), textcoords="offset points", xytext=(0,10), ha='center')

    # Set x-axis ticks to display all dates sequentially
    axs[0].set_xticks(stock_df['Date'])

    # Title and legend for the first subplot
    axs[0].set_title(f'Prev Closing Price for {stock_name}')
    axs[0].legend(loc='upper left')

    # Plotting another line chart in the second subplot
    # For example, let's plot the 'Haircut' column
    axs[1].plot(stock_df['Date'], stock_df['Haircut'], color='red', label='Haircut', marker='s')
    axs[1].set_xlabel('Date')
    axs[1].set_ylabel('Haircut', color='red')
    axs[1].tick_params('y', colors='red')

    # Rotate date labels for the second subplot
    axs[1].xaxis.set_major_formatter(DateFormatter('%d-%b'))
    axs[1].tick_params(axis='x', rotation=45)
    
    # Annotate each point with its y-value
    for i, (date, haircut) in enumerate(zip(stock_df['Date'], stock_df['Haircut'])):
        axs[1].annotate(f'{haircut:.2f}', (date, haircut), textcoords="offset points", xytext=(0,10), ha='center')

    # Set x-axis ticks to display all dates sequentially
    axs[1].set_xticks(stock_df['Date'])

    # Title and legend for the second subplot
    axs[1].set_title(f'Haircut for {stock_name}')
    axs[1].legend(loc='upper left')

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

# Take stock name from user input
stock_name = input("Enter stock name: ")

# Plot closing price and high for the given stock
plot_stock_closing_price(stock_name)
```
